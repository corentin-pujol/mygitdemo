## Import des modules pandas et numpy

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Import des données

```python
df=pd.read_csv("valeursfoncieres-2020.txt",sep="|")
```

```python
df.info()
```

```python
pd.DataFrame.head(df,n=10)
```

## Suppression des colonnes vides

```python
df2 = df.dropna(axis="columns",how="all")
pd.DataFrame.head(df2,n=1000)
```

# Différents types de logement

```python
df2['Type local'].value_counts()
```

# Histogramme

```python
df2['Type local'].value_counts().plot(kind='bar')
plt.show()
```

# Nature de la mutation 

```python
df2['Nature mutation'].value_counts()
```

# Histogramme

```python
df2['Nature mutation'].value_counts().plot(kind='bar')
plt.show()
```
